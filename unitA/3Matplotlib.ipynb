## OOP Style Using matplotlib

````python
%matplotlib inline
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12,6))
fig.suptitle('the figure')

ax1 = fig.add_subplot(2,2,1)
ax1.set_title('plot 1')
# plot on ax1.

ax2 = fig.add_subplot(2,2,2)
ax2.set_title('plot 2')
# plot on ax2.

ax3 = fig.add_subplot(2,2,3)
ax3.set_title('plot 3')
# plot on ax3.

ax4 = fig.add_subplot(2,2,4)
ax4.set_title('plot 4')
# plot on ax4.
````

````python
fig, axes = plt.subplots(2,2,figsize=(12,6))

fig.suptitle('the figure')

axes[0][0].set_title('plot 1')
# plot on axes[0][0]

axes[0][1].set_title('plot 2')
# plot on axes[0][1]

axes[1][0].set_title('plot 3')
# plot on axes[1][0]

axes[1][1].set_title('plot 4')
# plot on axes[1][1]
````

## A Basic Plot

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
x = np.arange(0.5,5.5,0.5)
y = x**2
plt.figure(figsize=(12,6))
plt.plot(x,y)  # simple function to be used for ploting
plt.title('First Plot')
plt.show()
````

````python
fig, ax = plt.subplots(1,1,figsize=(12,6))  # using normal way to create figure and plot
ax.set_title('First Plot')
ax.plot(x,y,'ro--')   # change the plot style
plt.show()
````

````python
fig, ax = plt.subplots(1,1,figsize=(12,6))
ax.set_title('First Plot',fontsize=28)
art1, art2, art3 = ax.plot(x,x,'bs-.',x,x**2,'ro--',x,np.log(x),'g^:')   # plot many.
ax.text(3,15,'$y=x^2$',fontsize=16,color='red')
ax.text(3.5,6,'$y=x$',fontsize=16,color='blue')
ax.text(4,0,'$y=log(x)$',fontsize=16,color='green')
ax.grid()
ax.legend([art2,art3,art1],['Series a, red','Series b, green','Series c, blue'],loc='upper center')
plt.show()
````

The format string is like this:  

    fmt = '[color][marker][line]'  

details of the format can look here the [docs from pyplot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)  
Aslo the plot title font size was changed.  
We also added text for each trend. The position of the text is given using the data coordinates.  
The grid() function toggle the grid show/off.

````python
x = np.arange(-2*np.pi,2*np.pi,0.01)
y = np.sin(3*x)/x
y2 = np.sin(2*x)/x
y3 = np.sin(x)/x

plt.figure(figsize=(12,6))
plt.plot(x,y,color='b')
plt.plot(x,y2,color='r')
plt.plot(x,y3,color='g')
plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi], [r'$-2\pi$',r'$-\pi$',r'$0$',r'$+\pi$',r'$+2\pi$'])
plt.yticks([-1,0,1,2,3], [r'$-1$',r'$0$',r'$+1$',r'$+2$',r'$+3$'])
plt.annotate(r'$\lim_{x\to 0}\frac{\sin(x)}{x}= 1$', xy=[0,1], xycoords='data', xytext=[1.2,1.5], fontsize=16, textcoords='data', arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

ax=plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))

plt.show()
````
