# Process data from HTML

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

read data from HTML page. Pandas will process all tables inside the page and make dataframe for each table and return the list of all dataframes.

```python
dfs = pd.read_html(r'https://en.wikipedia.org/wiki/List_of_cities_by_sunshine_duration')
len(dfs)
```

We can concat all the useful data into one dataframe and use it.

```python
df = pd.concat(dfs[1:],axis=0,keys=['Africa', 'Asia', 'Europe', 'North America', 'South America', 'Oceania'])
df
```

```python
def save_df_html(html_file, df):
    s = ['<HTML><HEAD><TITLE>']
    s.append('My DataFrame')
    s.append('</TITLE></HEAD><BODY>')
    s.append(df.to_html())
    s.append('</BODY></HTML>')
    html = ''.join(s)
    with open(html_file,'w',encoding='utf-8') as f:
        f.write(html)
    return

save_df_html('sunshine_data.html',df)
```
